```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 3.9, 6.2, 7.8, 9.5])

# MSE function
def mse(m):
    return np.mean((y - m*x)**2)

# Range of slope values
m_vals = np.linspace(0, 3, 300)
mse_vals = [mse(m) for m in m_vals]

plt.figure(figsize=(9,5))
plt.plot(m_vals, mse_vals, label="MSE(m)")
plt.title("MSE Curve (NO Gradient Descent)")
plt.xlabel("Slope m")
plt.ylabel("MSE")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Gradient function for MSE
def grad(m):
    n = len(x)
    return (-2/n) * np.sum(x * (y - m*x))

# Run gradient descent
m = 0.0
lr = 0.05
ms = [m]
losses = [mse(m)]

for _ in range(15):
    m = m - lr * grad(m)
    ms.append(m)
    losses.append(mse(m))

# Plot bowl + GD path
plt.figure(figsize=(9,5))
plt.plot(m_vals, mse_vals, label="MSE Curve")
plt.scatter(ms, losses, color="red", s=40, label="Gradient Descent Updates")
plt.title("Gradient Descent Following the MSE Bowl")
plt.xlabel("Slope m")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2.1, 3.9, 6.2, 7.8, 9.5])
n = len(x)

# Define MSE loss
def mse(m):
    return np.mean((y - m * x)**2)

# Define gradient
def grad(m):
    return (-2 / n) * np.sum(x * (y - m * x))

# Compute loss over range of slopes
m_vals = np.linspace(0, 3, 300)
mse_vals = [mse(m) for m in m_vals]

# Run gradient descent
m = 0.0
lr = 0.05
ms = [m]
losses = [mse(m)]

for _ in range(15):
    m = m - lr * grad(m)
    ms.append(m)
    losses.append(mse(m))

# Plot the loss curve
plt.figure(figsize=(9,5))
plt.plot(m_vals, mse_vals, label="MSE(m)")
plt.scatter(ms, losses, color="red", s=40, label="GD steps")
plt.xlabel("Slope m")
plt.ylabel("MSE Loss")
plt.title("MSE Loss Curve & Gradient Descent Path")
plt.legend()
plt.grid(True)
plt.show()
```
